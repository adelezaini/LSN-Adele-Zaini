## 2.1) Importance sampling

The aim of this exercise is computing the following integral using the Monte Carlo method:
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
and comparing two different samplings:

**a) Uniform sampling:** Random variables are extracted from the uniform ditribution in [0,1).


**b) Importance sampling** The idea is instead to extract variables from a pdf $p(x) \not= cost$. It can be shown that if $p(x)$ is properly chosen, the variance of the sampled integration is widely reduced.
 
 
<span style="color:purple">**<i>How to choose the proper $p(x)$?</i>**</span>

The pdf $p(x)$ has to be:

1. <i>similar to the integrand</i>.
2. positive: $p(x)\ge0 \ \ \forall x\in [0,1] $
3. normalized: $\int_{0}^{1}p(x)dx=1$

Usually the Taylor's expansion is a good choice, but in this case it does not satisfy all the conditions (it is not positive within the interval). So another idea is considering the straight line that meets the same boundary values of the integrand and then normalize it. The resulting pdf is $p(x)=2(1-x)$.

So we apply the importance sampling, considering a new integrand:

$$g(x)=\frac{\frac{\pi}{2}\cos(\pi x/2)}{2(1-x)}$$

and evaluating the indipendent variable $x$ as sampled from $p(x)$, through the cumulative inversion method:

$$F(x)=\int_{0}^{x} p(x') dx'=2(x-\frac{x^2}{2})=: y \rightarrow x=1-\sqrt{1-y}$$

where the (-) solution is taken to suit the right interval $[0,1]$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x1, y1, err1 = np.loadtxt("ex2-1/uniform.out", usecols=(0,1,2), unpack='true')
x2, y2, err2 = np.loadtxt("ex2-1/importance.out", usecols=(0,1,2), unpack='true')

plt.figure(figsize=(13, 10))
plt.title('Comparison between Uniform and Importance Sampling')
plt.errorbar(x1,y1,yerr=err1, color='y', label="Uniform Sampling")
plt.errorbar(x2,y2,yerr=err2, color='m', label="Importance Sampling")
plt.hlines(1, 0, 10000)
plt.xlabel('M')
plt.ylabel('Integral')
plt.grid(True)
plt.legend()
plt.show()
```

As we can see from the graph, the importance sampling makes the integral converging quickier to the actual solution, led by the <i>""micro""</i> uncertainties, i.e. square root of the variance.

## 2.2) Random walk

This exercise focuses on the Random Walk both in a 3D lattice and in the continuum. The first purpose is to evaluate $r_{RMS}=\sqrt{\langle |\vec{r}_s|^2 \rangle_{RW}}$, i.e. the quadratic mean of the distance from the origin; the second one is to check that it is actually a diffusive phenomenon and its law in function of the steps is: $r_{RMS}(s) \approx k\sqrt{s}$.

The $r_{RMS}$ is evaluated for $s=1,2,...100$ steps and the experiment is repeated $M=10^4$ times (divided into $N=100$ blocks).

<span style="color:purple">**<i>How to choose the next step?</i>**</span>

What changes between the two cases is the <i>"stepping rule"</i>:

**a) Discrete path:** The walker makes a forward or backward step of length $a=1$ in one of the three main directions $x$, $y$ or $z$.

To implement this condition, I chose to sample random variables uniformly in two different intervals:
1. in $[0,3)$ for the three directions: a) $x$ for $[0,1)$; b) $y$ for $[1,2)$; c) $z$ for $[2,3)$;

2. in $[0,2)$ for the orientation: a) forward for $[0,1)$; b) backward for $[1,2)$.

**b) Continuous path:** The walker makes a step of lenght $a=1$ in a random direction.

In order to compute the random direction we switch to polar coordinate system:

$$\begin{cases} x = a \ sin\theta \ cos\phi 
\\ y = a \ sin\theta \ sin\phi  
\\ z = a \ cos\theta \end{cases}$$

then the random direction is indicated by the solid angle, which is sampled by generating random value for $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$. 

We need to remember that the (normalized) jacobian for the coordinate change is $\frac{1}{2}\sin \theta $, so we sample $\theta$ according to the cumulative inversion method: $\theta=arcos(1-2y)$.

### a) Discrete path

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

def f (x, k):
    return k * np.sqrt(x)

x,y,err=np.loadtxt("ex2-2/discrete.out",usecols=(0,1,2),unpack=True)

fig=plt.figure(figsize=(15, 10))
plt.errorbar(x,y,yerr=err, label='Discrete path', linewidth=0.5, marker='.', color='darkblue',ecolor = 'darkblue')
plt.xlabel('Steps')
plt.ylabel('$r_{RMS}$')
plt.title("Discrete path", size='x-large', weight='bold')

#fitting
popt, pcov = curve_fit(f, x, y)
plt.plot(x,f(x,*popt), color='red', label='Fit')

plt.legend()
plt.grid(True)
plt.show()

print('The fit function is: f(s)=k √s with k=',popt[0])
```

### b) Continuous path

```python
x,y,err=np.loadtxt("ex2-2/continuous.out",usecols=(0,1,2),unpack=True)

fig=plt.figure(figsize=(15, 10))
plt.errorbar(x,y,yerr=err, label='Continuous path', linewidth=0.5, marker='.', color='darkblue',ecolor = 'blue')
plt.xlabel('Steps')
plt.ylabel('$r_{RMS}$')
plt.title("Continuous path", size='x-large', weight='bold')

#fitting
popt, pcov = curve_fit(f, x, y)
plt.plot(x,f(x,*popt), color='red', label='Fit')

plt.legend()
plt.grid(True)
plt.show()

print('The fit function is: f(s)=k √s with k=',popt[0])
```

As we can see in both graphs, there is a good compatibility between the sampling curve and its fitting $f(s)=k\sqrt{s}$.

This confirms that considering a large number $M$ of generated RW, this process can be considered diffusive, where $k=\sqrt{2D}$ and $D$ is the diffusive coefficient.
