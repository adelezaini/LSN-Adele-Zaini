# Exercises 4: Molecular Dynamics in the microcanonical NVE

The purpose of these exercises is improving a given code of Molecular Dynamics in a microcanonical ensemble NVE. The code firstly simulates the motion of interacting particles in a Lennard-Jones potential using the Verlet algorithm and then it evaluates the thermodynamic properties of the system.

### First of all, a bit of theory to understand...
**<i>Dynamics</i>**

The initial configuration of the simulation is the <i>fcc</i> crystal, so that we are sure not to have any overlap among particles positions. The time evolution of the configuration is evaluated thanks to the Verlet algorithm as follows:

$$\vec r(t+dt)=2\vec r(t)-\vec r(t-dt)+\frac{\vec F(t)}{m}dt^2$$

Note that it is not only needed the actual position, but also the previous one ( <i>"two-step method"</i> ).
The force acting among particles is computed from the energy potential ( $\vec F=-\vec \nabla V$ ).

**<i>Thermodynamic properties</i>**

<span style="color:purple">**Potential energy**</span>  The Lennard-Jones potential between each couple of particles is:

$$V_{ij}=4\epsilon\left[\left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12}-\left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{6}\right]$$

We apply a <i>spherical cutoff</i> $r_c$ in order to recude the number of calculations: since the largest contributions come from neighbour particles, it is not a bad approximation to neglet all interaction beyond $r_c$.

<span style="color:purple">**Kinetic energy**</span> It is computed by the definition: $K_i=\frac{1}{2}m(v_x^2+v_y^2+v_z^2)$
where velocities can be evaluated by the Verlet algorithm knowing consecutive positions.

<span style="color:purple">**Total energy**</span> It is simply the sum of the potential and the kinetic energy: $E=V+K$

<span style="color:purple">**Temperature**</span> It can be derive from the kinetic energy thanks to the <i>energy equipartion principle</i>: $T = 2K_{tot}/(3N)$

<span style="color:purple">**Pressure**</span> It can be estimated via the <i> virial theorem </i>:

$$ P = \rho k_B T + \frac{\left\langle W \right\rangle}{3V}$$
where:
$$ W =  \sum_{i=1}^{N-1} \sum_{j>i}^N 48\epsilon \left[ \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^{12} - 
\frac{1}{2} \left(\frac{\sigma}{|\vec{r}_i -\vec{r}_j|}\right)^6 \right] $$

**<i>Lennard-Jones units</i>**

It is often convenient to express quantities in reduced units to work with a dimensionless problem. In this case, we obtain the following relations with the properties expressed in SI unit:

$$ V= \epsilon V^* \qquad E= \epsilon E^* \qquad K= \epsilon K^* \qquad T = \frac{\epsilon}{k_B}T^* \qquad P = \frac{\epsilon}{\sigma^3} P^*$$

In this unit system the mass is considered as unity. Working in LJ units makes all the quantities of order 1 and equivalent systems with different combinations of their properties values can be treated the same.

**<i>Periodic Boundary Conditions</i>**

Since microscopically speaking the sample looks infinite, it is essential to reduce the spatial extension of the problem. The idea is considering a characteristic <i>3D box</i> that can be replicated infinitely times in the sample and contains a statistically significant number of particles without exceding in size. The sample is now characterized by a periodic system of same boxes. The important aspect concerns the particles that leave the considered box to enter a neighbour one: their images of the specular neighbour box are entering the present one so that the box density remains constant and the periodicity is conserved.

### 4.1) Equilibration: restart simulation and rescale velocities

The first improvement in the code is adding the possibility to restart the simulation from the final configuration of the previous one and to rescale velocities in order to match the input temperature. This is a fondamental requirement to equilibrate the system.
This is due to the fact that the code computes initial velocities randomly, but at each step collisions distribute energy among particles and velocities tend to assume a Maxwell distribution. What we see is the temperature dropping to lower values then the target one.
Restarting the simulation we "force" the temperature up again while progressively rescaling velocities by a factor: $ f_s= \sqrt{T_{target}/T_{actual}}=\sqrt{3 \ T_{target}/ \sum v^2}$.

To do so, I added two bool variables <i>restart</i> and <i>rescale</i> that can be change via the input file where the other parameters of the simulation are stored (<font face="Courier">input.dat</font>).

The steps to follow to manually equilibrate the system are:
1. start from the initial configuration (<i>fcc</i> configuration) and compute random velocities ($restart=0$, $rescale=0$) with a suggested number of steps $10^3$;
2. start from previous configuration and rescale velocities ($restart=1$, $rescale=1$). This last move can be repeated as many times as it is needed;
3. when equilibration is completed, increase the number of steps to $10^4$ and run <font face="Courier">bash clean.sh</font> in the shell to clean up all the output files, otherwise the results of the real simulation will be added to the previous ones.

**Note:** I prefered to mantain two separate commands for <i>restart</i> and <i>rescale</i> because they are actually two different actions (one could want to restart the simulation from the previous one without rescaling the velocities...) but one can quickly swap to the <i>restart+rescale</i> mode just changing <i>rescale</i> value ($restart=0$, $rescale=1$).

In order to check the effectiveness of the code improvements, I drew down two graphs showing the behaviour of the first four thermodinamic properties when restarting the simulation while rescaling velocities.

The parameter used to test the code are:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(12, 5))
y = np.loadtxt("ex4-3/Equilibration/output_temp.dat")
plt.plot(y, color='crimson')
plt.xlabel('# measurements')
plt.ylabel('T*')
plt.title('Temperature', color='crimson')
plt.hlines(1.1, 0, 800, color='midnightblue')
plt.grid(True)
plt.show()

fig=plt.figure(figsize=(12, 5))
y2 = np.loadtxt("ex4-3/Equilibration/output_ekin.dat")
y3 = np.loadtxt("ex4-3/Equilibration/output_epot.dat")
y4 = np.loadtxt("ex4-3/Equilibration/output_etot.dat")

plt.plot(y, color='maroon', label='Temperature')
plt.plot(y2, color='orangered', label='Kinetic energy')
plt.plot(y3, color='forestgreen', label='Potential energy')
plt.plot(y4, color='dodgerblue', label='Total energy')

plt.xlabel('# measurements')
plt.ylabel('E*')
plt.title('Energy', color='navy')
plt.hlines(1.65, 0, 800, color='sandybrown')
plt.hlines(-5, 0, 800, color='yellowgreen')
plt.hlines(-3.35, 0, 800, color='lightskyblue')
plt.legend()
plt.grid(True)
plt.show()
```

The temperature graph shows the value to drop as expected, but after 4/5 iterations the temperature tends to become stable around the target value.

The kinetic energy shows the same behaviour of the temperature (since they are the same quantity except for a proportional factor), i.e. the characteristic dropping that shows the tendency to distribute energy among particles.

An opposite behaviour is acquired by the potential energy that tends to conserve the total energy meanwhile the particle dispersion makes it descrease. The result of this opposite tendency is shown by the total energy trend that increases in each iteration until the system is equilibrated. This behaviour can be understood by thinking about the system as immersed in a heat bath at the start of each iteration: of course the system is provided of extra energy each time until the system temperature does not reach the heat bath temperature.

Now the three energies are quite stable and the total energy is now constant as expected in a microcanonical ensemble.

We can finally notice that they are all of order 1, so we are confident that the code is working well.

### 4.2) Average of thermodynamic properties

In this second exercise, we are asked to improve the code evaluating and printing not only the instant values but also the mean values and uncertaines of the thermodinamic properties in order to "smooth" the curve and show the global tendency while increasing the simulation steps.

We need to use the <i>data blocking method</i> in order not to have values too correlated to apply the <i> Central Limit Theorem </i>.


**Note:** it is important to add another filter to measurments within the same block otherwise the values are still too correlated and the error is too small. To do so, the thermodynamic properties are evaluated every 10 steps.

```python
fig=plt.figure(figsize=(15, 5))

x1,y1,yerr1 = np.loadtxt("ex4-3/Results/ave_temp.out", usecols=(0,1,2),unpack=True)
yt = np.loadtxt("ex4-3/Results/instant_temp.out")

plt.suptitle("Temperature", size='large', color='crimson')
plt.subplot(121)
plt.plot(yt, color='crimson')
plt.title('Instant value')
plt.hlines(1.1, 0, 1000, color='midnightblue')
plt.xlabel('# measurements')
plt.ylabel('T*')
plt.grid(True) 

plt.subplot(122)
plt.errorbar(x1, y1 ,yerr1, color='crimson')
plt.hlines(1.1, 0, 1000, color='midnightblue')
plt.title('Mean value')
plt.xlabel('# measurements')
plt.ylabel('T*')
plt.grid(True) 

plt.show()

fig=plt.figure(figsize=(15, 5))

x2,y2,yerr2 = np.loadtxt("ex4-3/Results/ave_ekin.out", usecols=(0,1,2),unpack=True)
x3,y3,yerr3 = np.loadtxt("ex4-3/Results/ave_epot.out", usecols=(0,1,2),unpack=True)
x4,y4,yerr4 = np.loadtxt("ex4-3/Results/ave_etot.out", usecols=(0,1,2),unpack=True)

yk = np.loadtxt("ex4-3/Results/instant_ekin.out")
yp = np.loadtxt("ex4-3/Results/instant_epot.out")
ye = np.loadtxt("ex4-3/Results/instant_etot.out")

plt.errorbar(x1, y1, yerr1, color='crimson', label='Temperature')
plt.errorbar(x2, y2, yerr2, color='orangered', label='Kinetic energy')
plt.errorbar(x3, y3, yerr3, color='forestgreen', label='Potential energy')
plt.errorbar(x4, y4, yerr4, color='dodgerblue', label='Total energy')

plt.plot(yt, color='grey')
plt.plot(yk, color='grey')
plt.plot(yp, color='grey')
plt.plot(ye, color='grey')

plt.title("Mean thermodynamic variables", size='x-large', color='navy')
plt.xlabel('# measurements')
plt.ylabel('E*/T*')
plt.hlines(1.65, 0, 1000, color='sandybrown')
plt.hlines(-5, 0, 1000, color='yellowgreen')
plt.hlines(-3.35, 0, 1000, color='lightskyblue')
plt.legend()
plt.grid(True)
plt.plot()
plt.show()
```

The actual mean values match very well the mean values I previously evaluated manually.
