```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(12, 5))
y = np.loadtxt("Equilibration/output_temp.dat")
plt.plot(y, color='crimson')
plt.xlabel('# steps')
plt.ylabel('T*')
plt.title('Temperature', color='crimson')
plt.hlines(1.1, 0, 800, color='midnightblue')
plt.grid(True)
plt.plot()
plt.show()

fig=plt.figure(figsize=(12, 5))
y2 = np.loadtxt("Equilibration/output_ekin.dat")
y3 = np.loadtxt("Equilibration/output_epot.dat")
y4 = np.loadtxt("Equilibration/output_etot.dat")

plt.plot(y, color='maroon', label='Temperature')
plt.plot(y2, color='orangered', label='Kinetic energy')
plt.plot(y3, color='forestgreen', label='Potential energy')
plt.plot(y4, color='dodgerblue', label='Total energy')

plt.xlabel('# steps')
plt.ylabel('E*')
plt.title('Energy', color='navy')
plt.hlines(1.65, 0, 800, color='sandybrown')
plt.hlines(-5, 0, 800, color='yellowgreen')
plt.hlines(-3.35, 0, 800, color='lightskyblue')
plt.legend()
plt.grid(True)
plt.plot()
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [14, 5]
fig, ax = plt.subplots(1,2)

x1,y1,yerr1 = np.loadtxt("ave_press.out", usecols=(0,1,2), delimiter=' ', unpack='true')
y2 = np.loadtxt("instant_press.out", usecols=(0), delimiter=' ', unpack='true')
ax[1].plot(y2)
ax[0].errorbar(x1, y1 ,yerr1)
ax[0].set_xlabel('#block')
ax[0].set_ylabel('Temp')
ax[0].set_title('Temperature mean value')
ax[1].set_xlabel('#steps')
ax[1].set_ylabel('Temp')
ax[1].set_title('Temperature istant value')
ax[0].grid(True)
ax[1].grid(True)
ax[0].plot([0, 1000], [1.1, 1.1], color='k', linestyle='-', linewidth=1.5)
ax[1].plot([0, 1000], [1.1, 1.1], color='k', linestyle='-', linewidth=1.5)
plt.show()
```
