# Exercises 10: Optimization techniques

Last time we employed the Genetic Algorithm as optimization technique to explore the global optimum of the Travelling Salesman Problem. This time we are asked to implement two other metheuristic methods: the **simulated annealing** and the **parallel tempeting**.

## 10.1) Simulated Annealing

### SA algorithm

As the GA was based on a biological process, here the analogy between the algorithm and a natural phenomenon lies in the thermodynamics: finding the global optimum is compared to a liquid/metal involved in a **slow** cooling process until the configuration is crystallized, the temperature is at $0 \ K$ and the energy reaches the minimum value (i.e. ground state).

The key point lies in the interpretation of the cost function as the total energy of the system $E=L(x)$ that has to be minimized to reach the ground state. In this scheme, the whole process results linear rather than 'clustered' as the GA:
0. Generate the initial configuration $x=x_0$ and evaluate the cost function $L(x_0)$ (i.e. the 'square lenght' of the tour);
1. At a given $T_i$ (or $\beta _i$ ), repeat the following steps $n_i$ times to reach the configuration $x_i$:
    - Propose the next configuration $x'$ applying mutations to $x$;
    - Evaluate the quality of the new proposed move throught its cost function $L(x')$:
        - if $L(x')<L(x)$, $x'$ is an improved solution so directly accepted;
        - if $L(x')>L(x)$, it can still be accepted considering the Boltzmann distribution, sampled with the Metropolis Algorithm. 
        
        In other words, the probability to accept the proposed move is:
        
    $$ p(x)=\begin{cases}
        e^{-\beta _i \ (L(x')-L(x))} & \text{if $L(x')>L(x)$}.\\
        1 & \text{otherwise}.
          \end{cases}$$
    
        Note that the probability decreases exponentially with the “badness” of the move, which is related to the amount $\Delta E$ by which the solution is worsened (i.e. energy is increased).


2. Lower the temperature $T_i \rightarrow T_{i+1}$ and repeat the previous step, where the initial input configuration is $x_i$.

In order to simulate the cooling of a liquid/metal down to its freezing point, we need to set the so called **annealing schedule**, which is a discrete and finite sequence of temperature values decreasing progressively: $\{(n_1,T_1), (n_2,T_2), \dots , (n_N, T_N)\}$, where the index $i$ refers to the point 1 of the previous list, $T_i$ is the relative temperature and $n_i$ the number of times that has to be repeated (different for each temperature).

To sum up the logical sequence is:
$$ \begin{cases}
    x_0 \rightarrow \left [\text{Metropolis}\ (T_1) \right ]^{n_1}\rightarrow x_1 \\ 
    T_1\rightarrow T_2 \\ 
    x_1\rightarrow \left [\text{Metropolis}\ (T_2) \right ]^{n_2} \rightarrow x_2 \\ \dots \\
    T_{N-1}\rightarrow T_N \\ 
    x_{N-1}\rightarrow \left [\text{Metropolis}\ (T_N) \right ]^{n_N} \rightarrow x_N \\ 
\end{cases}$$

### SA implementation
The purpose of the first exercise is to implement the Simulate Annealing algorithm in the case of the Travelling Salesman Problem. Previous codes from Metropolis simulations and from the GA are merged and readapted: the basic scheme is the one of a Metropolis simulation (initial configuration $\rightarrow$ next step proposed $\rightarrow$ accepted or rejected? $\rightarrow$ restart iteration), but now we work on the class Tour as the configuration $x$ previously mentioned, that can mutate and then be accepted or rejected. Each Metropolis sampling reprents a point of the annealing schedule.

<span style="color:purple">**<i>How to calibrate parameters for the annealing schedule?</i>**</span>

The calibration of the parameters in the annealing schedule is crucial to the algorithm performance, since the probability to select a move depends directly on temperature. What one can note is that at high temperatures the acceptance rate of a propose move is higher (particles moves freely and can assume several configurations), while it decreases with lowering the temperature (particles progressively lose mobility and assume configurations more and more similar to the ground state). It is very important to perform a **slow** cooling, allowing redistribution of the particles as they lose mobility, otherwise the system does not preserve the property of ergodicity and the algorithm does not converge to the global optimum.

First of all I decided to apply an <i> exponential cooling scheme </i>, instead of a linear one (i.e. constant decrement move): $T_t=T_0\ e^{\ -(1- \alpha)\ t}$ so that we can assume $ T_{i+1}= \alpha T_{i}$ with $\alpha <1$. In this way we "skim" high temperatures whereas we can better explore values close to the freezing point. Also, we never face the possibility to go out of the boundary limit of $T=0$, since this method is asimptotical. The choice of $\alpha$ influences the velocity of approching the freezing point. To conserve ergodicity, it is then very important to choose it properly.

Other paramenters to be optimized are the numbers of iterations $ \{n_i \}$ of the Metropolis sampling. A simple choice could be set them all equal to only one value s.t. $n=n_i \ \forall \ i$. Nevertheless, we can note that $T_i$ and $n_i$ are correlated if we want to preserve both ergodicity and a good algorithm efficiency: we aim to visit all the space of configurations, while focusing on the global minimum. For this reason, I decided to set smaller values for initial $n_i$ and then increase it while time passing. Given the target number $n_{final}$, the very first $n_0$ is chosen empirically as: $n_0= 2\ n_{final}/ T_0$; then it is incremented of 100 moves each time temperature is increased until the value reaches $n_{final}$. Note that in this case I chose a linear scheme because now I just want to impose a condition on the first temperature-steps.
So deciding to diversificate $n_i$ we can then start from a higher temperature, while containing the computational effort.

Running the simulation a few times, I set the following parameters that can be easily changed via the file <font face="Courier">input.dat</font>:

- Initial temperature $T_0=20$;
- Time steps $100$;
- Temperature rate $\alpha=0.9$;
- Target number of iterations $n_{final}=10^4$;

where the time steps (or temperature steps) are chosen considering the precision of approaching the freezing point.

### a) Best tour lenght $L^{(2)}$ in function of temperature

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.gridspec as gridspec

def upload(filename):
    n, x, y = np.loadtxt(filename, usecols=(0, 1,2), unpack=True) 
    x1,y1= np.loadtxt(filename, usecols=(1,2), max_rows=1,unpack=True)
    newx=np.append(x,x1)
    newy=np.append(y,y1)
    return n,newx,newy
```

```python
fig = plt.figure(figsize=(15,7))
plt.suptitle("Evolution of $L^{(2)}$ – SA", weight='bold', size='x-large', color='darkmagenta')

fig = plt.subplot(121)
g, t, b = np.loadtxt("SA/Circonference/Best_Lenght.out", usecols=(0,1,2), unpack=True)
#t=np.flip(t)
plt.plot(t, b, color='purple')
plt.title('Circonference')
plt.xlabel('Temperature (log)')
plt.ylabel('$L^{(2)}$')
plt.xscale("log")
plt.grid(True)
plt.xlim([10**(1.5),10**(-3.5)])

fig = plt.subplot(122)
g, t, b = np.loadtxt("SA/Square/Best_Lenght.out", usecols=(0,1, 2), unpack=True)
#t=np.flip(t)
plt.plot(t, b, color='purple')
plt.title('Square')
plt.xlabel('Temperature (log)')
plt.ylabel('$L^{(2)}$')
plt.xscale("log")
plt.grid(True)
plt.xlim([10**(1.5),10**(-3.5)])

plt.show()
```

### b) Best tour

```python
fig= plt.figure(figsize=(15,6.5))
plt.suptitle("Best Tour – SA", weight='bold', size='x-large', color='darkmagenta')


gridspec.GridSpec(1,7)

for t in range(2):
    if t==0: 
        filename0="SA/Circonference/Positions.out"
        filename="SA/Circonference/Best_Tour.out"
        plt.subplot2grid((1,7), (0,0), colspan=4, rowspan=4)
    else:
        filename0="SA/Square/Positions.out"
        filename="SA/Square/Best_Tour.out"
        plt.subplot2grid((1,7), (0,4), colspan=4, rowspan=3)
        
    n0, x0, y0 = upload(filename0)
    n, x, y = upload(filename)
    ncities=len(x0)

    plt.plot(x0,y0, color='darkblue', alpha=0.2, label='Initial tour')
    plt.plot(x,y, color='darkblue', label='Final tour')
    plt.scatter(x,y, marker='*', s=150, c=np.arange(ncities), cmap='gnuplot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    
    if t==0:
        plt.ylim(-1.3,1.3)
        plt.xlim(-1.3,1.3)
        plt.title('Positions of cities - Circonference')
        cbar = plt.colorbar()
        cbar.set_label('final order of visit')
        plt.clim(vmin=1,vmax=ncities)
    else:
        plt.ylim(-0.7,0.7)
        plt.xlim(-0.7,0.7)
        plt.title('Positions of cities - Square')

plt.show()
```

As we can see from the previous pictures, the simulation results hit the expectations converging to an optimized configuration.

We can notice that, as in the previous case of the GA, the convergence in the circonference case is quickier than in the square one. One can easily guess why: even though cities in both cases are randomly chosen in a 2D space, the rule for the circonference it is actually of placing the cities **on a** (curved) **line**, while selecting cities placed **inside** a square their positions are more chaotic and then it is harder to find the optimized path.

## 10.2) Parallel Tempering

The basic idea of PT is to perform the same simulation in parallel with multiple nodes, changing input data, in order to better explore the space of configurations and the evolution of the algorithm. The purpose is to reach the global optimum avoiding to be trapped in some metastable configuration.
One is free to choose how to improve this basic principle: making systems comunicate, swapping similar sistems, changing the input temperature.... 

In the following exercise we are asked to run the previous Simulated Annealing simulation indipendently on multiple nodes and compare each performance at the end. We start from the same initial configuration in order to make the single node behaviour comparable with each other, but changing the initialization of the <i>Random variable</i> of each node (selecting different lines of the <font face="Courier">Primes</font> file). The purpose is to look if the global minimum is reached quickier and more accurately than before, since there are "five minds" working on the same problem.

In particular, I decided to leave the code as untouched as possible so the single node would perform exactly the same run as the previous exercise (printing all the best tour configurations and lenghts); at the end I just implemented a function <i>FindBestest()</i>, which uses the subroutine <font face="Courier">MPI_Reduce</font> to find the minimum value and its location (i.e. rank) and the results were printed in the rispective output files.

Here are the results of running five nodes in parallel.

### a) Best tours lenght $L^{(2)}$ in function of temperature

```python
fig = plt.figure(figsize=(15,12))
plt.suptitle("Evolution of $L^{(2)}$ – PT", weight='bold', size='x-large', color='darkmagenta')

size=5;
colors=['deeppink','darkviolet','mediumslateblue','dodgerblue','mediumturquoise','limegreen','seagreen','darkgreen','navy']

fig = plt.subplot(211)

for i in range(size):
    g, t, b = np.loadtxt("PT/Circonference/Best_Lenght_"+str(i)+".out", usecols=(0,1,2), unpack=True)
    plt.plot(t, b, color=colors[i], alpha=0.75, label='Process n.'+str(i))
    
plt.title('Circonference')
plt.xlabel('Temperature (log)')
plt.ylabel('$L^{(2)}$')
plt.xscale("log")
plt.grid(True)
plt.legend()
plt.xlim([10**(1.5),10**(-3.5)])

fig = plt.subplot(212)
for i in range(size):
    g, t, b = np.loadtxt("PT/Circonference/Best_Lenght_"+str(i)+".out", usecols=(0,1,2), unpack=True)
    plt.plot(t, b, color=colors[i], alpha=0.75, label='Process n.'+str(i))
    
plt.title('Square')
plt.xlabel('Temperature (log)')
plt.ylabel('$L^{(2)}$')
plt.xscale("log")
plt.legend()
plt.grid(True)
plt.xlim([10**(1.5),10**(-3.5)])

plt.show()
```

### b) Best tours

```python

#fig= plt.figure(figsize=(15,6.5))
#gridspec.GridSpec(1,7)

plt.rcParams['figure.figsize'] = [13, 30]
fig, ax = plt.subplots(size,2)
plt.suptitle("Best tours – PT", weight='bold', size='x-large', color='darkmagenta')

for t in range(2):    
    for i in range(size):
        if t==0: 
            filename="PT/Circonference/Best_Tour_"+str(i)+".out"
            ax[i][t].set_ylim(-1.3,1.3)
            ax[i][t].set_xlim(-1.3,1.3)
            ax[i][t].set_title('Positions of cities - Circonference')
        else:
            filename="PT/Square/Best_Tour_"+str(i)+".out"
            ax[i][t].set_ylim(-0.7,0.7)
            ax[i][t].set_xlim(-0.7,0.7)
            ax[i][t].set_title('Positions of cities - Square')
        n, x, y = upload(filename)
        ncities=len(x)
        ax[i][t].plot(x, y, color=colors[i], alpha=0.75, label='Process n.'+str(i))
        ax[i][t].scatter(x,y, marker='*', s=150, c=np.arange(ncities), cmap='cool')

        ax[i][t].set_xlabel('x')
        ax[i][t].set_ylabel('y')
        ax[i][t].legend()

plt.show()
```

We can see that all processes have a good convergence and give the same final result in both cases. The only differences we can notice are the direction of walking (clockwise or anticlockwise) and the starting point of the tour (vivid pink). 

We can now guess which is the 'bestest' path.

### c) Bestest

```python
def upload2(filename, row):
    init=np.loadtxt(filename, usecols=(0), unpack=True, max_rows=row)
    n, x, y = np.loadtxt(filename, usecols=(0, 1,2), skiprows=row, unpack=True) 
    x1,y1= np.loadtxt(filename, usecols=(1,2), skiprows=row, max_rows=row+1,unpack=True)
    newx=np.append(x,x1)
    newy=np.append(y,y1)
    return init, n,newx,newy
```

```python
fig= plt.figure(figsize=(15,6.5))
plt.suptitle("Bestest tour – PT", weight='bold', size='x-large', color='darkmagenta')
gridspec.GridSpec(1,7)

for t in range(2):
    if t==0: 
        filename="PT/Circonference/Bestest_Tour.out"
        plt.subplot2grid((1,7), (0,0), colspan=4, rowspan=4)
    else:
        filename="PT/Square/Bestest_Tour.out"
        plt.subplot2grid((1,7), (0,4), colspan=4, rowspan=3)
        
    rank, n, x, y = upload2(filename,1)
    ncities=len(x)

    plt.plot(x,y, color='darkblue', label='Final tour')
    plt.scatter(x,y, marker='*', s=150, c=np.arange(ncities), cmap='gnuplot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    
    if t==0:
        plt.ylim(-1.3,1.3)
        plt.xlim(-1.3,1.3)
        plt.title('Positions of cities - Circonference')
        cbar = plt.colorbar()
        cbar.set_label('final order of visit')
        plt.clim(vmin=1,vmax=ncities)
    else:
        plt.ylim(-0.7,0.7)
        plt.xlim(-0.7,0.7)
        plt.title('Positions of cities - Square')

plt.show()
```
