
If we compare the evolution of the algorithm as hill-climbing, the annealing schedule determines the degree of uphill movement and is, thus, critical to the algorithm's performance: at higher $T$, uphill moves are more likely to occur, while as $T$ tends to zero, they become more and more unlikely (see the previous probability). This means that 

The principle underlying the choice of a suitable annealing schedule lies in the analogy with the metal/liquid behaviour: the initial temperature $T_0$ should be high enough to <i>melt</i> the system completely, then should be progressively reduced towards its <i>freezing point</i>  $T_N=0 K$ (?), 

Con questo schema di lavoro si riesce a bilanciare entrambe le richieste di ergodicità e di minimizzazione dell'energia. Partendo da alte temperature e con una diminuzione molto lenta il sistema è infatti libero di "visitare" molte configurazioni senza incastrarsi in un minimo locale.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.gridspec as gridspec

def upload(filename):
    n, x, y = np.loadtxt(filename, usecols=(0, 1,2), unpack=True) 
    x1,y1= np.loadtxt(filename, usecols=(1,2), max_rows=1,unpack=True)
    newx=np.append(x,x1)
    newy=np.append(y,y1)
    return n,newx,newy

def upload2(filename, row):
    init=np.loadtxt(filename, usecols=(0), unpack=True, max_rows=row)
    n, x, y = np.loadtxt(filename, usecols=(0, 1,2), skiprows=row, unpack=True) 
    x1,y1= np.loadtxt(filename, usecols=(1,2), skiprows=row, max_rows=row+1,unpack=True)
    newx=np.append(x,x1)
    newy=np.append(y,y1)
    return init, n,newx,newy
```
