# Exercises 5: Hydrogen atom

The purpose of this exercise is implementing the Metropolis algorithm and applying it to a quantum mechanics problem: the Hydrogen atom.

### Metropolis algorithm

The Metropolis algorithm is a Monte Carlo method used to approximate probability distributions or to compute integrals. It is very powerful because can sample **any** probability density in **any** number of dimensions. 

The hypothesis of its application relies on the system that has to be ergodic and so the **stochastic transfer probability $K(y|x)$** (i.e. the kinetics) has to be regular and has to satisfy the detailed balance. According to these statements, the transfer probability can be considered as product of a **trial transition probability** and an **acceptance probability**: $K(y|x)=T(y|x) \times A(y|x)$.

If the trial transition distribution is choosen symmetric, the acceptance probability can be evaluate as: $A(y|x)=\text{min} [1, p(y)/p(x)]$.

The very first move of the algorithm is the <i>Initialization</i>: choose the starting point $x_0$ and the calibrated proposal distribution T(y|x). The recursive steps that have to be iterated $m=1,2....,M$ times are:
1. Generate candidate $x'$ for the next sample picking from $T(x'|x_m)$;
2. Evaluate the acceptance probability $\alpha := A(x'|x_m)$ with the previous formula;
3. Accept or reject?
    - Generate $r$ from a uniform distribution in [0,1];
    - <i>Accept</i>:   if $r \leq \alpha$ $\ \ \ \rightarrow \ x_{m+1}=x'$;
    - <i>Reject</i>:   if $r > \alpha$ $\ \ \ \ \rightarrow \ x_{m+1}=x_m$.
    
Of course, the choice of trial distribution $T(y|x)$ is essential to the efficiency of the algorithm, as I explain later while calibration parameters such as the time step.

### Hydrogen atom
In quantum physics, each particle state is described in terms of probability functions, called **wave functions $\Psi(\vec x)$**. They depends on parameters called **quantum numbers**, with represent the symmetries of the system and each combinantion of them characterize a different particle state and so a different wave function.

Interesting and analitically simple wave functions $\Psi_{n,l,m}(r,\theta,\phi)$ are the ones refered to the ground state ( <i>1s </i> ) and one of the first excited state ( <i>2p </i>) of the electrons in a Hydrogen atom:

$$\Psi_{1s}=\Psi_{1,0,0}(r,\theta,\phi)=\frac{a_0^{3/2}}{\sqrt{\pi}} e^{-r/a_0 } $$

$$\Psi_{2pz}=\Psi_{2,1,0}(r,\theta,\phi)=\frac{a_0^{5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0 } \cos(\theta)$$

where $a_0=4\pi\epsilon_0\hbar^2/m_e e^2=0.0529 \ nm$ is the <i>Bohr radius</i>, $n=1,2,...$ is the <i>principal quantum number</i>, $l=0,1, ..., n-1$ the <i>azimuthal quantum number</i>, $m=-l, -l+1, ..., l-1, l$ the m<i>agnetic quantum number</i>.

Many behaviours and features of the system can be derived from operations on these functions, considering that $|\Psi(\vec x)|^2$ represents the probability density of the system. 

For example, we may want to evaluate the mean value of the radius of the electron orbital in a Hydrogen atom:

$$\left\langle r \right\rangle_{\Psi_{n,l,m}} = \int r \ |\Psi_{n,l,m}(r, \theta, \phi)|^2 \ d \vec x $$

The choice of the Hydrogen atom is due to the semplicity of the analytical expression of the previous wave functions, that makes possible the exact calculation of the results in order to check, learn and calibrate the Metropolis algorithm:

Expectation values for the radius:
$$\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0\quad\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0 $$


### Exercise
We were asked to implement the Metropolis algorithm in order to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ using a uniform and gaussian transition probability $T(\vec{x}|\vec{y})$ and then evaluate the mean radius for each orbital using data blocking, verifying their convergence to the actual values. In order to work on a dimensionlesss problem, we work in Bohr radius unit sistem: $r/a_{0} \rightarrow r$.

After a picture of the results, I will discuss more deeply the calibration of the simulation parameters.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig=plt.figure(figsize=(15, 5))

xu, yu, erroru = np.loadtxt("Uniform/Radius_1s.out", usecols=(0,1,2), unpack=True)
xg, yg, errorg = np.loadtxt("Gaussian/Radius_1s.out", usecols=(0,1,2), unpack=True)
plt.errorbar(xu,yu,yerr=erroru ,color='crimson', label="Uniform" )
plt.errorbar(xg,yg,yerr=errorg ,color='forestgreen', label="Gaussian")
plt.title("Mean radius of 1s orbital")
plt.xlabel('N')
plt.ylabel('$<r>/a_0$')
plt.hlines(1.5, 0, 200, color='midnightblue')
plt.grid(True)
plt.legend()
plt.show()

fig=plt.figure(figsize=(15, 5))
xu, yu, erroru = np.loadtxt("Uniform/Radius_2p.out", usecols=(0,1,2), unpack=True)
xg, yg, errorg = np.loadtxt("Gaussian/Radius_2p.out", usecols=(0,1,2), unpack=True)
plt.errorbar(xu,yu,yerr=erroru ,color='crimson', label="Uniform")
plt.errorbar(xg,yg,yerr=errorg ,color='forestgreen', label="Gaussian")
plt.title("Mean radius of 2p orbital")
plt.xlabel('N')
plt.ylabel('$<r>/a_0$')
plt.hlines(5, 0, 200, color='midnightblue')
plt.grid(True)
plt.legend()

plt.show()
```

We can notice that the gaussian choice for $T(y|x)$ makes the mean radius trend <i>adhere</i> to the real solution quickier and more accurately than the uniform one. We could expect this result: given a $r$ value from the previous iteration that has been accepted and we generate the next step close to the previous with more probability than far from it, the process results to be more efficient.

## Calibration of parameters
### Origin of the simulation and equilibration
Since Metropolis algorithm samples a probability distribution that asymptotically tends to the desired one $p(\vec x)$, we can consider the sequence of steps as a random walk of the $\vec x$ variable <i>"fluctating around"</i> the maximum(s) of the target pdf $p(\vec x)$. So we expect that if we start from far from $\vec x_{max}$ (s.t. $p(\vec x_{max})=p_{max})$, after some steps the random walk tends to converge around $\vec x_{max}$ (i.e. mean value of sampled steps).

In order to equilibrate the sistem with a few steps, I expect that for 1s orbital the best choise of starting position is $\vec x_{0}= \vec x_{max}=(0,0,0)$ since it corresponds to a wave function with spherical simmetry around the origin of the reference axis.

In contrast, the 2p orbital chosen (with quantum numbers $n=2, \ l=1, \ m=0$) shows two maximums simmetric to the plan $xy$ (at height $z=0$). So a good choise could be starting around one of the two, setting $x_{0}=0$ and $y_{0}=0$ because these symmetries are still conserved.

In order to verify my expectations and to find the best $z_{0}$ for the 2p orbital, I made a quick simulation of $M=10^3$ steps to focus only on the first steps of the random walk to look at the trend line of each coordinate ($x,y,z$) of each orbital (1s, 2p) using both transition distribution (Uniform, Gaussian), with the time step value optimized as explain later. Then, I found the mean value of the step sequence without considering the first part of equilibration.

Here I show an example picture of the calibration starting from $\vec x_{0}=(10,10,10)$ on a) $x$ - 1s - Uniform;  b) $z$ - 2p - Gaussian:

```python
fig=plt.figure(figsize=(12, 10))

x= np.loadtxt("Uniform/Origin/Positions_1s.out", usecols=(0), unpack=True, max_rows = 400)
xm=np.loadtxt("Uniform/Origin/Positions_1s.out", usecols=(0), unpack=True, skiprows=150, max_rows = 1000)

z = np.loadtxt("Gaussian/Origin/Positions_2p.out", usecols=(2), unpack=True, max_rows = 1000)
zm=np.loadtxt("Gaussian/Origin/Positions_2p.out", usecols=(2), unpack=True, skiprows=150, max_rows = 1000)

plt.suptitle("Origin calibration", size='large', color='crimson')
plt.subplot(211)
plt.plot(x, color='crimson')
plt.title('x - 1s - Uniform')
plt.hlines(np.mean(xm), 0, 400, color='midnightblue')
plt.xlabel('# steps')
plt.ylabel('$x/a_0$')
plt.grid(True) 

plt.subplot(212)
plt.plot(z, color='crimson')
plt.title('z - 2p - Gaussian')
plt.hlines(np.mean(zm), 0, 1000, color='midnightblue')
plt.xlabel('# steps')
plt.ylabel('$z/a_0$') 
plt.grid(True)

print ("x_0 for the 1s orbital with Uniform pdf is approximately: \n", np.mean(xm), "\n")
print ("z_0 for the 2p orbital with Gaussian pdf is approximately: \n", np.mean(zm))

plt.show()
```

**Observations:**
- The random walk of 1s orbital behaves as expected: quickly converging to 0. Same behaviour for the other two spatial variables (not shown).
- For 2p orbital, it is evident that $M=10^3$ steps are few to sample both maximums, but it is a good choice to find the mean value of the <i>"upper"</i> $z_{max}$, so that $\vec x_0=(0,0,4)$. Note that any $z$ value s.t. $0<z<8$ is frequently visited.
- From this method, varying the starting point $\vec x_0$, it was possible to find a good number of steps to equilibrate the system: it is proportional to the order of magnitude of chosen $x_0$, $y_0$, $z_0$. I chose $10$ as the factor of proportionality to be sure to overcome the equilibration time. (Note: with such an accurate choice of the starting point this first part could be little useful, but I decided to keep it to mantain the program general and able to equilibrate from any starting point).

### Time step
The choice of the time step $\delta $ is very important for the quality of the results of the Metropolis algorithm: large moves cause a low efficiency while small ones can affect data with high correlation. It is a common rule to look for the $\delta$ s.t. the algorithm gives an accettance rate of $ \approx 50 \% $ ( i.e. the <i>"50% empirical rule"</i>).

Even though I could implement a function to do it authomatically, I prefered to calibrate $\delta$ manually just changing its value from the input file because it is a very quick and accurate method. The values I found are:
- for uniform distribution: 

$ \delta (1s) = 1.226 $ with $e_A \approx 10^{-3}$

$ \delta (2p) = 2.975 $ with $e_A \approx 10^{-4}$

- for gaussian distribution: 

$ \delta (1s)= 0.761 $ with $e_A \approx 10^{-5}$

$ \delta (2p) = 1.878 $ with $e_A \approx 10^{-5}$

where $e_A$ represents the relative error of the acceptance rate compared to the target value 0.5: $e_A=|A-0.5|/0.5$.

### Number of blocks

Since the number of steps is highier than usual, I decided to increase also the number of blocks. Running the program a few times changing $N$ and looking at the error bars width of $<r>$, I considered $N=200$: the error bars do not look too small, indicator of high correlation among data, while decreasing $N$ the error bars do not change in width (hypothesis of Central Limit Theorem are satisfied).
