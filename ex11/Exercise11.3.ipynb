## 11.3) Function of two variables

Finally we are asked to perform our model to fit a trigonometric 2D function: 

$$f(x,y) = \sin(x^2+y^2)$$ 

in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

### a) Dataset

```python
#libraries
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

def f(x, y):
    return np.sin(x**2+y**2)

def printscore(score):
    print('******************************************************')
    print('Model performance:')
    print('– Test loss:', score[0])
    print('– Test accuracy:', score[1])
    print('******************************************************')
```

```python
n_valid=2000
n_train=10*n_valid
n_epochs=150
sigma=0.2

low=-3/2
high=3/2
    
# generate training inputs and validation inputs
np.random.seed(0)

xy_train=np.zeros((n_train,2))
xy_train[:,0]=np.random.uniform(low, high, n_train) 
xy_train[:,1]=np.random.uniform(low, high, n_train)

xy_valid=np.zeros((n_valid,2))
xy_valid[:,0]=np.random.uniform(low, high, n_valid) 
xy_valid[:,1]=np.random.uniform(low, high, n_valid)

z_target=f(xy_valid[:,0],xy_valid[:,1]) 

# actual measures from which we want to guess regression parameters
z_train= np.random.normal(f(xy_train[:,0], xy_train[:,1]), sigma) 
z_valid= np.random.normal(f(xy_valid[:,0], xy_valid[:,1]), sigma)

#3D representation
X = np.copy(xy_valid[:,0])
Y = np.copy(xy_valid[:,1])
X.sort()
Y.sort()
X, Y = np.meshgrid(X,Y)
Z=f(X,Y)

fig = plt.figure(figsize=(10, 8))
plt.suptitle("Dataset")
ax = Axes3D(fig)

ax.scatter(xy_valid[:,0], xy_valid[:,1], z_valid, c=z_valid, cmap='winter', label='Test')
ax.plot_surface(X, Y ,Z, cmap='PuBu', linewidth=0, antialiased=True, alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```

### b) Model

Since in the previous exercise, model 'D' showed the best results I trained a model with the same choice of parameters of the previous exercise.

| $n$     | $HL$    |    $F$  |   $O$   |   $C$   |
| ------- | ------- | ------- | ------- | ------- |
|   30    |    3    |   tanh  |  Adam   | logcosh |

```python
#parameters
n_neurons=30
n_layers=3
act='tanh'
opt='Adam'
los='logcosh'

model = tf.keras.Sequential()
model.add(Dense(n_neurons, input_shape=(2,), activation=act))

for i in range(n_layers):
    model.add(Dense(n_neurons, activation=act))

model.add(Dense(1))

model.compile(optimizer=opt, loss=los, metrics=['mse'])
#plot_estimate = PlotCurrentEstimate(x_valid, y_valid)
#earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=100, mode='auto')
history= model.fit(xy_valid, z_valid, batch_size=32, epochs=n_epochs, validation_data=(xy_valid, z_valid), verbose=0) #callbacks=[ plot_estimate, earlystop],

fig = plt.figure(figsize=(13,10))

#model.summary()
# evaluate model
score = model.evaluate(xy_valid, z_valid, batch_size=32, verbose=0)
printscore(score)
par=model.count_params()
print("Total number of parameters: ", par)
print('*******************************************************')

#predicted value
xy_predicted=np.zeros((n_valid,2))
xy_predicted[:,0]=np.random.uniform(low, high, n_valid) 
xy_predicted[:,1]=np.random.uniform(low, high, n_valid)

z_predicted = model.predict(xy_predicted)

plt.suptitle("Predicted data")
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, cmap='GnBu', linewidth=0, antialiased=False, alpha=0.3)

ax.scatter(xy_predicted[:,0], xy_predicted[:,1], z_predicted, c='navy', label='Predicted', marker='*', s=50) #c = np.abs(z_valid), cmap='gnuplot',


ax.set_xlabel('x')
ax.set_ylabel('y')
plt.legend()
plt.show()
```

The results confirm that the parameters were well calibrated and the model performs very well in the given range.
