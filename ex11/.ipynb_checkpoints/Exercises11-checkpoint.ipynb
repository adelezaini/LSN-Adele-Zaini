## <span style="color:steelblue">Supervised Learning</span>

The idea of Supervised Learning is to <i>train</i> the model on given data already <i>"labelled"</i> in order to optimize the parameters of the model (i.e. weights $w_i$ and bias $b$). 

The process is divided into a <i>training</i> and a <i>testing</i> schedule, as follows:
<div> <img src="supervised.JPG" width="650"/></div>

In order to evaluate the performance of the model a <i>cost function</i> $C [ Y, f(X,p) ] $ is needed, where $Y$ represents the labelled input data (alias expected output) and $f(X,p)$ the model output as function of the input $X$ (alias actual input) and the parameter $p$. The model is fit finding the parameters that minimize the cost function.

Operatively speaking, the very first step is to randomly divide the dataset $D(X,Y)$ into two groups $D_{train}$ and $D_{test}$ (the latter should be about $10 \%$ of the first). The optimization of the parameters is applied to the train dataset: $p= \text{argmin}_p \{ E_{in} \}$, where $E_{in}$ is the <i>in-sample error</i> $E_{in}=C [ Y_{train}, f(X_{train},p) ] $. Finally the evaluation of the perfomarce is found using the <i> out-of-sample error </i> on the test set: $E_{out}=C [ Y_{test}, f(X_{test},p) ]$ .

It is important to note that the purpose of the whole process is to minimize $E_{out}$ (not $E_{in}$! otherwise we face overfitting) in function both of the data points and the model complexity: while in the first case the dependence is monotonic and decreasing with increasing number of data points, the latter shows a minimum in intermediate complexity. This means that according to the available dataset (usually it is fixed, otherwise we choose the maximum possible amount), we need to choose an appropriate model structure without exceding in complexity.

## <span style="color:chocolate">Exercises</span>
The Supervised Learning is particularly suitable for classification and regression problems. In the following exercises we are asked to implement a NN in order to optimize the parameters to better fit the given functions $f(x)$ (linear, polynomial, multivariable).

### <span style="color:brown">Dataset</span>
The dataset $D(x,y)$ is represented by the indipendent variables $x_i$ and the dipendent $y_i$ generated starting from the exact values $f(x_i)$ as follows:
$$ y_i= f(x_i) + \eta_i $$

where $\eta_i$ is a Gaussian noise variable such that $ \langle \eta_i \rangle=0$ and $ \langle \eta_i \eta_j \rangle = \delta_{ij} \sigma$.

### <span style="color:brown">Cost function</span>
For continous function, an appropriate choice for the cost function is the mean-square-error between the expected $ \{y_i\}$ and the predicted $\{y_i^{pred}\}$ s.t. $y_i^{pred}=F(x_i)$, where I re-named $F(x_i)$ as the model product to avoid misunderstanding:

$$ MSE= \frac{1}{N_\mathrm{test}}\sum_{j=1}^{N_\mathrm{test}} (y_j^\mathrm{pred}-y_j)^2$$
### <span style="color:brown">Optimizer</span>
In order to find the minimum of the cost function we need to use optimization tecniques. Here we will use the <i> Stocastic Gradient Descent </i> applied to the NN thanks to the <i> Backpropagation algorithm</i>. After computing the gradient of the loss function with respect to the weights for a single input–output run, it updates the weights by a tiny amount in the direction of a better solution. It is then necessary to run it several times to converge to the optimum of the cost function.

### <span style="color:brown">Metrics</span>
Running the training phase several times to reach to minimum of the cost function, it is then needed a "cost function for the cost function". This is called <i>metrics</i>. It is then a particular cost function that judges the model performance and it is not used directly when training the model. Here we will use the mean-square-error $MSE$.

## 10.1) Linear fit

Given the function expression:
$$ f(x)=2x+1 $$

and a prototypical code, we are asked to improve the code varying the number of epochs, $N_{epochs}$, the number of data points $N_{train}$ and the noise $\sigma$.

```python
#libraries
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
from tensorflow.keras.models import load_model
```

### a) Dataset
I increased the number of training (and test) points by a factor of $10$, while I chose $\sigma=0.3$ to make data more stochastic.

```python
#target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

# generate training and test inputs
np.random.seed(0)
n_valid=100
x_train = np.random.uniform(-1, 1, n_valid*10)
x_valid = np.random.uniform(-1, 1, n_valid)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.3 #cambia anche sopra!
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)

# plot training and target dataset
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='pink', marker='*', alpha=0.2, label='Train')
plt.plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2')
plt.scatter(x_valid, y_valid, color='orange', marker='*', label='Test')
plt.title("Dataset")
plt.legend()
plt.grid(True)
plt.show()
```

### b) Neural network structure
In order to have two parameters to optimize (slope $m$ and intersect $b$) we just need a single neuron in a single hidden layer.

```python
# compose the NN model
model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,))) 

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

### c) Training phase
As I previously mentioned the training phase needs to be run several times before reaching the optimum of the cost function. An **epoch** is a single run in which the training dataset is passed through the neural network. After the first epoch, we will be provided by the first couple of proposed parameters. Of course, a single epoch leads to underfit the function. As the number of epochs increases, more number of times the weight are changed in the neural network and the curve goes from underfitting to optimal to overfitting curve. 

Since there is no rule of the number of epochs, we need to run the following script a few times to give extimation of the proper numebr of epochs. I finally chose to leave $N_{epochs}=30$ because an increase of this number would not lead to better results.

Each epoch is comprised of one or more batches. A **banch** is a subgroup of training data after which the parameters are updated. This additional operative layer is needed to avoid overfitting and to reach more efficiently and more accurately the minimum of the cost function. The <font face="Courier">banch_size</font> is the number of training data points in each batch and it is conventionally set as power of $2$.

```python
# fit the model using training dataset
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid), verbose=0)
```

### d) Optimized parameters

```python
# return weights
optm, optb = model.get_weights()
print("Optimized m = ", optm[0][0])
print("Optimized b = ", optb[0])
```

### e) Performance evaluation over validation and exact data

```python
# evaluate model
score1 = model.evaluate(x_valid, y_valid, batch_size=32, verbose=2)
# evaluate model with the exact curve
score2 = model.evaluate(x_valid, y_target, batch_size=32, verbose=2)
```

```python
# print performance
print()
print('"Test" loss:', score1[0])
print('"Test" accuracy:', score1[1])

print()
print('"Exact" loss:', score2[0])
print('"Exact" accuracy:', score2[1])
```

### f) In-sample and Out-of-sample errors

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['$E_{in}$', '$E_{out}$'], loc='best')
plt.show()
```

### g) Optimized model at work

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)

plt.figure(figsize=(10, 6))
plt.plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2', alpha=0.75)
plt.scatter(x_predicted, y_predicted, color='crimson', marker='*')
plt.title("Predicted data")
plt.legend()
plt.grid(True)
plt.show()
```

Varying the parameters I could notice that:

- fixed initial $N_{train}$ and $\sigma>0$, $N_{epochs}$ had to be increased to $\approx 100$ to get appreciable results;
- fixed initial $N_{epochs}$ and $\sigma>0$, $N_{train}$ had to reach $\approx 80$.
- the noise $\sigma$ affects little the performance of the model.

Exploring other combinations of this parameters, I noticed that increasing the number of training value leads to more accurate results than changing the number of epochs, which instead cause time consuming.
