## 11.2) Polynomial fit

The purpose of the second exercise to extend the model to obtain a reasonable fit of the following polynomial function:

$$f(x)=4-3x-2x^2+3x^3$$

for $x \in [-1,1]$. The parameters to calibrate are:

- the number of layers $N_L$
- the number of neurons in each layer $n$
- the activation function $F(z)$
- the optimizer $O$
- the cost function $C$

### Dataset

I decided to increase the number of epochs $N_{epochs}=100$, the train datapoints $N_{train}=2000$, the test datapoints $N_{test}=200$ and to reduce a bit the noise dispersion $\sigma=0.3$ since the function to fit is more complex than the previous. 

```python
#libraries
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
from tensorflow.keras.models import load_model

def f(x):
    return 4-3*x-2*x**2+3*x**3

def printscore(score):
    print('*******************************************************')
    print('Model performance:')
    print('– Test loss:', score[0])
    print('– Test accuracy:', score[1])
    print('*******************************************************')
```

```python
n_valid=200
n_train=10*n_valid
n_epochs=100
sigma=0.3

# generate training and test inputs
np.random.seed(0)

x_train = np.random.uniform(-1, 1, n_train)
x_valid = np.random.uniform(-1, 1, n_valid)
x_valid.sort()
y_target = f(x_valid)

y_train = np.random.normal(f(x_train), sigma) 
y_valid = np.random.normal(f(x_valid), sigma)

plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='pink', marker='*', alpha=0.2, label='Train')
plt.plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2')
plt.scatter(x_valid, y_valid, color='orange', marker='*', label='Test')
plt.title("Dataset")
plt.legend()
plt.grid(True)
plt.show()
```

### Calibration of parameters

###  <span style="color:darkmagenta"> a) Number of layers $N_L$ and neurons in each layer $n$ </span> 

In order to understand how the NN behaves with different structures, I used a Jupyter Notebook (<font face="Courier">Training_11-2</font> in the folder 'Models-11.2') running the NN several times changing the parameters and saving each found model. Here are comparisons varying just one or two parameters at a time while the others are fixed.

The thre frist pictures show the dependance to the number of hidden layers $HL$ and neurons in each layer $n$, choosing $F(z)=selu$, $O=sgr$ and $C=mse$. 

```python
plt.rcParams['figure.figsize'] = [17, 6]
fig, ax = plt.subplots(1,2)

markers=["*","o","8", "s", "p", "^", "D", "X"]
colors=['deeppink','violet','darkviolet','mediumslateblue','dodgerblue','limegreen','seagreen','darkgreen','navy']

x_predicted = np.random.uniform(-1, 1, 150)

for s in range(2):
    if(s==0): I=6
    else: I=7
    for i in range(I):
        if(s==0):
            n_layers=[1,2,3,5,10,15]
            filename='Models_11-2/layers_'+str(n_layers[i])
            labelname='Hidden layers = '+str(n_layers[i])
            title="Predicted data: different HL"
        if(s==1):
            n_neurons=[1,5,10,20,50,100,200]
            filename='Models_11-2/neurons_'+str(n_neurons[i])
            labelname='Neurons = '+str(n_neurons[i])
            title="Predicted data: different neurons"
        model=load_model(filename)
        y_predicted = model.predict(x_predicted)
        ax[s].scatter(x_predicted, y_predicted, s=20, color=colors[i], marker=markers[i], label=labelname)
        x_predicted.sort()
        if (i>0):
            ax[s].plot(x_predicted, y_predicted, color=colors[i], alpha=0.5)

    ax[s].plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2', alpha=0.75)
    ax[s].set_title(title)
    ax[s].legend()
    ax[s].grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
n_neurons=[10,50,100]
n_layers=[3,5]

x_predicted = np.random.uniform(-1, 1, 150)

for n in range(4):
    for l in range(2):
        i=l+n*2;
        if(i<6):
            filename='Models_11-2/neurons_'+str(n_neurons[n])+'-layers_'+str(n_layers[l])
        elif(i==6):
            filename='Models_11-2/neurons_'+str(10)+'-layers_'+str(10)
        else:
            break
        model=load_model(filename)
        y_predicted = model.predict(x_predicted)
        if(i<6):
            plt.scatter(x_predicted, y_predicted, s=30, color=colors[i], marker=markers[i], label='n= '+str(n_neurons[n])+', HL='+str(n_layers[l]))
        elif(i==6):
             plt.scatter(x_predicted, y_predicted, s=30, color=colors[i], marker=markers[i], label='n= '+str(10)+', HL='+str(10))
        
        x_predicted.sort()
        if(i>0):
            plt.plot(x_predicted, y_predicted, color=colors[i], alpha=0.5)
        
plt.plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2', alpha=0.75)
plt.title("Predicted data: combinations of HL e neurons")
plt.legend()
plt.grid(True)
plt.show()
```

We can then conclude that it is sufficient to increase a little the number of neurons for each layer that we obtain remarkable results, whereas changing the number of layers leads to little improvements in the model.

### <span style="color:darkmagenta"> b) Activation function $F(z)$ </span> 
We can notice from the following graphs that choosing 'tanh' or 'selu' lead to the best results.

```python
plt.rcParams['figure.figsize'] = [17, 6]
fig, ax = plt.subplots(1,3)
act=['sigmoid', 'tanh', 'elu', 'relu', 'selu']
num=[1,5,10]

x_predicted = np.random.uniform(-1, 1, 150)

for n in range(3):
    for a in range(5):
        filename='Models_11-2/'+str(num[n])+'-'+str(num[n])+'-'+act[a]
        model=load_model(filename)
        y_predicted = model.predict(x_predicted)
        ax[n].scatter(x_predicted, y_predicted, s=20, color=colors[a], marker=markers[a], label=act[a])
        x_predicted.sort()
        if (a>0):
            ax[n].plot(x_predicted, y_predicted, color=colors[a], alpha=0.5)
        
    ax[n].plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2', alpha=0.75)
    ax[n].set_title("Predicted data: n=HL="+str(num[n]))
    ax[n].legend()
    ax[n].grid(True)
plt.show()
```

###  <span style="color:darkmagenta"> c) Optimizer $O$ </span> 

The best optimizer is always the 'sgd', but also 'RMSprop' and 'Adam' can give remarkable results.

```python
plt.rcParams['figure.figsize'] = [17, 6]
fig, ax = plt.subplots(1,3)
opt=['sgd', 'RMSprop', 'Adagrad', 'Adam']
num=[1,5,10]

x_predicted = np.random.uniform(-1, 1, 150)

for n in range(3):
    for a in range(4):
        filename='Models_11-2/'+str(num[n])+'-'+str(num[n])+'-'+opt[a]
        model=load_model(filename)
        y_predicted = model.predict(x_predicted)
        ax[n].scatter(x_predicted, y_predicted, s=20, color=colors[a], marker=markers[a], label=opt[a])
        x_predicted.sort()
        if (a>0):
            ax[n].plot(x_predicted, y_predicted, color=colors[a], alpha=0.5)
        
    ax[n].plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2', alpha=0.75)
    ax[n].set_title("Predicted data: n=HL="+str(num[n]))
    ax[n].legend()
    ax[n].grid(True)
plt.show()
```

### <span style="color:darkmagenta">d) Cost function $C$ </span> 

Choosing 'mse' leads always to the best parametrization of the model, but increasing the number of neurons and hidden layers even the 'logcosh' could be a good choice.

```python
plt.rcParams['figure.figsize'] = [17, 6]
fig, ax = plt.subplots(1,3)
los=['mse','mae','msle', 'logcosh', 'poisson']
num=[1,5,10]

x_predicted = np.random.uniform(-1, 1, 150)

for n in range(3):
    for a in range(5):
        filename='Models_11-2/'+str(num[n])+'-'+str(num[n])+'-'+los[a]
        model=load_model(filename)
        y_predicted = model.predict(x_predicted)
        ax[n].scatter(x_predicted, y_predicted, s=20, color=colors[a], marker=markers[a], label=los[a])
        x_predicted.sort()
        if (a>0):
            ax[n].plot(x_predicted, y_predicted, color=colors[a], alpha=0.5)
        
    ax[n].plot(x_valid, y_target, color='darkorchid', label="Target", linewidth='2', alpha=0.75)
    ax[n].set_title("Predicted data: n=HL="+str(num[n]))
    ax[n].legend()
    ax[n].grid(True)
plt.show()
```

### Calibrated models

Here are some models with different combinations of the parameters that led to the best results.



| Model   | $n$     | $HL$    | $F$  |     $O$ |     $C$ |
| ------- | ------- | ------- | ------- | ------- | ------- |
| A       | 50      |       3 |    tanh |     sgd |     mse |
| B       | 50      |       5 |    selu |    Adam |     mse |
| C       | 20      |      10 |    tanh |     sgd | logcosh |
| D       | 30      |       3 |    tanh |    Adam | logcosh |


```python
plt.rcParams['figure.figsize'] = [17, 12]
fig, ax = plt.subplots(2,1)
n_model=['A','B','C','D']
title=["inside","outside"]

x_predicted = np.random.uniform(-2, 2, 300)
x_out=np.arange(-2, 2, 0.1)
y_out=f(x_out)
score=[]
par=[]

for s in range(2):
    for i in range(4):
        model=load_model('Models_11-2/model'+n_model[i])
        y_predicted = model.predict(x_predicted)
        ax[s].scatter(x_predicted, y_predicted, s=30, color=colors[i], marker=markers[i], label='Model '+n_model[i])
        #x_predicted.sort()
        #ax[s].plot(x_predicted, y_predicted, color=colors[i], alpha=0.5)
        score.append(model.evaluate(x_valid, y_valid, batch_size=32, verbose=0))
        par.append(model.count_params())

    ax[s].plot(x_out, y_out, color='darkorchid', label="Target", linewidth='2', alpha=0.75)
    ax[s].set_title("Predicted data "+title[s]+" the range")
    ax[s].legend()
    ax[s].grid(True)
    if(s==0): 
        ax[s].set_xlim(-1,1)
        ax[s].set_ylim(1.4,5)
    else: 
        ax[s].set_xlim(-2,2)
        ax[s].set_ylim(-15,10)
plt.show()

for i in range(4):
    print(n_model[i])
    printscore(score[i])
    print("Total number of parameters: ", par[i])
    print('*******************************************************')
```

We can then observe two facts:
- model 'D' is the one that shows the best performance, but it is also the one with least paramenters involved. So taking into count also the 'Model complexity' results we can than conclude that a more complex model is not a well performing model as well. It is necessary to find a proper balance of numbers of parameters to reach the best quality in results.
- from the last graph, we understand how the Machine Leaning works. It is very different from usual processes where we already have the model structure (i.e. in the regression it is the function expression) where we just need to calibrate the parameters of the formulae (e.g. slope, intersect...). Here the model is <i>trained</i> and <i>learns</i> thanks to the given dataset (both input and output) to give a complete model as a result. This is why we have so many parameters (i.e. weights and bias) and why the model gives appropriate results only where it is trained.
