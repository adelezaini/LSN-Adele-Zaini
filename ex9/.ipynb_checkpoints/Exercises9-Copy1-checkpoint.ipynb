# Exercises 9: Travelling Salesman Problem

The Travelling Salesman Problem represents a classical example of a Genetic Algorithm application.

Given a set of cities randomly distributed (on a 2D space), the aim of the salesman is visiting all the cities only once with the shortest possible round-tour.

Every possible tour is then classified on a <i>'fitness'</i> basis, related to its (square) lenght:
$$L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}|^2$$
where $\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N$ are the positions of cities in the order they are visited and s.t. $\vec{x}_{N+1}=\vec{x}_1$.

## Genetic Algorithm
The Genetic Algorithm, as the name suggests, is based on the phenomen of the <i>natural selection</i> among the individuals of a spiece according to their ability to adapt. Generation after generation, only the most adaptable ones survive and reproduce.

The GA is then based on the following steps:
0. Create the **initial population** with random individuals;
1. Evaluate the **fitness** of each individual and so the probability to be selected.
2. Produce the **next generation**:

    a. select a parent;
    
    b. with probability $P_c$, **crossover**: a part of its genes will be merge with genes of another selected parent to produce an offspring.
    
    c. if crossover doesn't occur, then copy the parent in the next generation;
    
    d. with probability $P_m$, **mutation**: the offspring can face some unpredictable changes;
    
    
3. The last step is repeated until the new generation doesn't fullfil the fitness conditions.

## Exercise

In order to implement this idea in the TSP case, <i>object oriented programming</i> is applied with the creation of the two following classes:
- <i>**Tour**</i>: it is basically a vector of cities with some methods implemented to make some operations more efficient (e.g. evaluating its lenght, swapping two elements...). It is based on the idea of creating a one-to-one correspance between the city position and its index in order to work simply with a vector of integers (e.g. $\left[ 1, 2, 3, 4, 5 \right]$ where $1 \leftrightarrow (x_1,y_1)\dots$).
- <i>**Population**</i>: it is instead a vector of tours, which is able to execute all the operations just mentioned (initialization, selection, crossover, mutation, sorting...). Note that a) for the selection operator the Russian Roulette method ha been applied; b) the mutation is randomly choosen within four possibilities (swap of two random elements, total shift of $n$ positions, one group shift of $n$ positions, invertion); c) the crossover has a particular application (the 'mum part' is copied, while the second part is composed by the missing elements in the order they appear in the 'dad'); d) in order to increase efficiency but to avoid to be trapped in some local minimum, after $50$ iterations the <i>elitism</i> is activated (i.e. directly copying the best ones to the next generation before the selection phase).

For further information, you can look at the header file <font face="Courier">TSP.h</font>. 

The number of cities, their position configuration, the number of population members, the number of generations and the probabilities of crossover and mutation are all contained in an input file (<font face="Courier">input.dat</font>) to be quickly changed to study some properties of the algorithm. 

Here results are shown setting the following parameters:
- Number of cities $n=30$;
- Number of individuals in each population $n^2=900$;
- Number of generations $g=300$;
- Probability of crossover $P_c=55\%$;
- Probability of mutation $P_m=9\%$;

in two possible configurations, where the cities are randomly placed:
- on circonference of $r=1$;
- within a square of $l=1$.

### a) Initial configuration
### Circonference

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.gridspec as gridspec

n, x, y = np.loadtxt("Circonference/Positions.out", usecols=(0, 1,2), unpack=True)
ncities=len(x)

x1,y1= np.loadtxt("Circonference/Positions.out", usecols=(1,2), skiprows=1, max_rows=2,unpack=True)
newx=np.append(x,x1)
newy=np.append(y,y1)

fig= plt.figure(figsize=(15,7))
gridspec.GridSpec(1,3)

plt.subplot2grid((1,3), (0,0), colspan=2, rowspan=3)
plt.plot(newx,newy, color='darkblue', alpha=0.2)
#plt.subplot(121)
plt.scatter(x,y, marker='*', s=150, c=np.arange(ncities), cmap='gnuplot')#'RdPu','viridis'
plt.xlabel('x')
plt.ylabel('y')
plt.title('Positions of cities')
plt.ylim(-1.3,1.3)
plt.xlim(-1.3,1.3)

cbar = plt.colorbar()
cbar.set_label('order of visit')
plt.clim(vmin=1,vmax=ncities)

plt.subplot2grid((1,3), (0,2))
#plt.subplot(122)
columns = ('City Index', 'x', 'y')
#rowslab = ['    %d    ' % i for i in n]
plt.axis('tight')
plt.axis('off')

data=[]

for i in range(ncities):
    data.append(["%d" % (n[i]+1), "%.3f" % x[i], "%.3f" % y[i]])

    
table= plt.table(cellText=data,colLabels=columns,loc='center', colColours=['orange', 'orange', 'orange'])

plt.show()
```
